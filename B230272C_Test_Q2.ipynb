```python
import pandas as pd

spotify = pd.read_csv("spotify.csv")
df = pd.DataFrame(spotify)
df
```

```python
print (df.columns)
```

```python
df_2017 = df[df['track_album_release_date'].str.startswith('2017')]
df_2017
```

```python
grouped_df = df_2017.groupby('playlist_genre').agg({
    'loudness': 'mean',
    'mode': 'mean',
    'speechiness': 'mean',
    'acousticness': 'mean',
    'instrumentalness': 'mean',
    'liveness': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'duration_ms': 'mean'
}).reset_index()

grouped_df['duration'] = grouped_df['duration_ms'] / 60000
grouped_df.drop(columns='duration_ms', inplace=True)

grouped_df
```

```python
grouped_df.to_csv('spotify_2017.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lmplot(x='danceability', y='energy', data=df_2017, aspect=2, height=5)
plt.title('Relationship between Danceability and Energy (2017)')
plt.show()
```

```python
correlation = df_2017[['danceability', 'energy']].corr()
print("Correlation between danceability and energy:", correlation)
```

```python
numerical_df = df_2017.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python
#find the correlation between other factors to danceability and energy
correlation_danceability = correlation_matrix[['danceability']]
correlation_energy = correlation_matrix[['energy']]

print(correlation_danceability)
print(correlation_energy)
```
